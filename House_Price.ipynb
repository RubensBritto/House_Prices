## Import the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Import the .cvs

```python
link_test_csv = 'https://raw.githubusercontent.com/RubensBritto/House_Prices/main/Data/train.csv'
test_csv = pd.read_csv(link_test_csv)
test_csv
```

## Analysing the Data Set



### Analysing the columns

```python
test_csv.info()
```

### Analyzing the amounts of the NaN

```python
#test_csv['Alley'].isnull().sum()
#test_csv.info()
#test_csv.isna()
valuesNaN = [test_csv[i].isnull().sum() for i in test_csv if test_csv[i].isnull().sum()!=0]
labels = [i for i in test_csv if test_csv[i].isnull().sum()!=0]
print([f'{i} : {j}'for i,j in zip(labels,valuesNaN)])
```

### Creating the graph to visualize the NaN values

```python
fig = plt.figure(figsize = (20, 10))
plt.barh(labels, valuesNaN) 
plt.ylabel("Labels") 
plt.xlabel("Null Amount")  
plt.show() 
```
